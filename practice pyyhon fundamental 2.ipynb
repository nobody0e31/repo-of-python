```python
#question no 8
a = int(input("Enter a number: "))
b= int(input("Enter b number: "))

def calculator(a,b,operation):
    if operation == "+":
        return a+b
    elif operation == "-":
        return a-b
    elif operation == "*":
        return a*b
    elif operation == "/":
        return a/b
    else:
        print("Invalid operation")
    
operation = input("Enter the operation (+, -, *, /): ")
result = calculator(a,b,operation)
print("The result is: ", result)
```

```python


def is_prime(n):
    if n < 2:
        return False
    
    i = 2
   
    
    while i < n:
        if n % i == 0:
            return False
        i = i + 1
    return True

user_input = int(input("Enter the number: "))

# Pass the integer into the function and store the returned True/False
result = is_prime(user_input)

# Print the final boolean result
print(result)
```
